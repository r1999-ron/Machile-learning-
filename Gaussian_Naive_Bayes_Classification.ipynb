```
# usual imports 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
seaborn.set()

plt.figure(figsize=(10,8))
#creating the data map 
from sklearn.datasets import make_blobs
X, y = make_blobs(100,2,centers = 2,cluster_std=1.5,random_state = 2)
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap="RdBu")
```

```
#getting the naive bayes classifier
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X,y)  #fitting the model 

#generationg new data 
rng = np.random.RandomState(0)
X_test = [-6,-14] + [14,18] *rng.rand(2000,2)
y_preds = clf.predict(X_test) #predicting for the new data 
Y_preds_prob = clf.predict_proba(X_test)
```

```
#visualizing the predictions inorder to figure out the decision boundaries
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap="RdBu")
lim = plt.axis()
plt.scatter(X_test[:,0],X_test[:,1],c=y_preds,s=1000,cmap="RdBu",alpha=0.05)
plt.axis(lim)
```
