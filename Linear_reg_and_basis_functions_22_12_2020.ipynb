```
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

#Simple Linear Regression

```
#create the data 
rng = np.random.RandomState(1)
X = 10*rng.rand(50) 
y = 2*X - 5 + rng.randn(50)

#plotting the data
plt.figure(figsize=(6,4))
plt.scatter(X,y)
plt.xlabel(" X ",color = "white")
plt.ylabel(" y ",color= "white")
```

```
#using the scikit learn linear reg
from sklearn.linear_model import LinearRegression
reg1 = LinearRegression(fit_intercept = True)
#fitting
reg1.fit(X.reshape(-1,1),y)

#using another set to test and pull out the plot 
x_test = np.linspace(0,10,1000)
y_test = reg1.predict(x_test.reshape(-1,1))

#plotting the final regression line 
plt.scatter(X,y)
plt.scatter(x_test,y_test,c="r",linewidths=0.1,alpha=0.08)
print("model slope: ",reg1.coef_[0])
print("model intercept :",reg1.intercept_)
```

```
#testing with multi dimentional data 
rng = np.random.RandomState(1)
X = 10* rng.rand(100,3) #random values in a 3x1 matrix
y = 0.5 + np.dot(X,[1.5,2,2.5]) # dot products the coef to the X values and adds the intecept 
#fitting the same linear model
reg1.fit(X,y)

#checking for cslope and intercept 
print("model slope: ",reg1.coef_)
print("model intercept :",reg1.intercept_)
```
