#Robust Regression 

```
x = df["RM"].values.reshape(-1,1)
y = df["MEDV"].values

#building the regressor 
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor()
ransac.fit(x,y)
```

```
# creating the best fit line with the predictions 
line_x = np.arange(3,10,1)
line_y = ransac.predict(line_x.reshape(-1,1))
```

```
# we are going to note the outliers => create the inlier and the oulier mask 
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask) #stating that whatever is not the inlier mask should be the outlier 
#sns.set(style="darkgrid",context="notebook")
plt.figure(figsize=(10,6))
plt.scatter(x[inlier_mask],y[inlier_mask],color ="blue",label="inliers")
plt.scatter(x[outlier_mask],y[outlier_mask],color="red",label="outliers")
plt.plot(line_x,line_y,color="brown")
plt.xlabel("no of rooms(avg)")
plt.ylabel("median house value")
plt.legend(loc="upper left")
```

```
#the predictions :
predictions = pd.DataFrame(line_y,line_x)
predictions.columns = ["predicted label"]
predictions
```

```
from sklearn.model_selection import train_test_split
x = df.drop("MEDV",axis=1).values
y=df["MEDV"].values
x_train, x_test, y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
```

```
lin.fit(x_train,y_train)
y_train_preds = lin.predict(x_train)
y_test_preds = lin.predict(x_test)
```

#Residual Analysis

```
plt.figure(figsize=(10,6))
plt.scatter(y_train_preds,y_train_preds-y_train,color = "blue",label="training predictions")
plt.scatter(y_test_preds,y_test_preds-y_test,color = "orange",label ="test predictions")
plt.xlabel("predictions")
plt.ylabel("Residuals")
plt.legend(loc = "upper left")
plt.hlines(y=0,xmin=-10,xmax=50,lw=1)
plt.xlim([-10,50])
```

```
#mean squarred error 
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train,y_train_preds) , mean_squared_error(y_test,y_test_preds)
```

coeff of determination 


```
from sklearn.metrics import r2_score
r2_score(y_train,y_train_preds) , r2_score(y_test,y_test_preds)
```
